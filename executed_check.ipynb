# Phase 2: Data Preprocessing & Feature Engineering

Create Working Copy

```python
df_processed = df.copy()

print(df_processed.shape)
```

Check Missing Values Again

```python
missing = pd.DataFrame({
    'Missing_Count': df_processed.isnull().sum(),
    'Missing_Percentage':
        (df_processed.isnull().sum()/len(df_processed))*100
})

missing.sort_values(
    by='Missing_Percentage',
    ascending=False
)
```

Handle Missing Values

```python
num_cols = df_processed.select_dtypes(
    include=['int64','float64']
).columns

for col in num_cols:
    df_processed[col].fillna(
        df_processed[col].median(),
        inplace=True
    )
```

```python
cat_cols = df_processed.select_dtypes(
    include=['object']
).columns

for col in cat_cols:
    df_processed[col].fillna(
        df_processed[col].mode()[0],
        inplace=True
    )
```

```python
df_processed.isnull().sum().sum()
```

Remove Duplicates

```python
before = df_processed.shape[0]

df_processed.drop_duplicates(
    inplace=True
)

after = df_processed.shape[0]

print("Removed:", before-after)
```

```python
risk_mapping = {
    'Low':0,
    'Medium':1,
    'High':2
}

df_processed['Layoff_Risk'] = (
    df_processed['Layoff_Risk']
    .map(risk_mapping)
)
```

# Phase 3: Machine Learning Models

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

import pandas as pd
import numpy as np
```

Evaluation Function

```python
from sklearn.ensemble import VotingClassifier
import numpy as np

def evaluate_model(model, X_test, y_test):
    X_test_processed = X_test

    if isinstance(model, VotingClassifier):
        individual_predictions = []
        print("\n--- VotingClassifier Individual Estimator Predictions ---")

        for name, estimator in model.named_estimators_.items():
            pred = np.asarray(estimator.predict(X_test_processed)).reshape(-1)
            individual_predictions.append(pred)
            print(f"Estimator '{name}' prediction shape: {pred.shape}")

        print("------------------------------------------\n")

        if model.voting == 'hard':
            encoded_predictions = np.vstack([
                model.le_.transform(pred)
                for pred in individual_predictions
            ]).T

            weights = getattr(model, '_weights_not_none', None)
            majority_votes = np.apply_along_axis(
                lambda row: np.argmax(
                    np.bincount(
                        row.astype(int),
                        weights=weights,
                        minlength=len(model.le_.classes_)
                    )
                ),
                axis=1,
                arr=encoded_predictions
            )
            y_pred = model.le_.inverse_transform(majority_votes)
        else:
            y_pred = np.asarray(model.predict(X_test_processed)).reshape(-1)
    else:
        y_pred = np.asarray(model.predict(X_test_processed)).reshape(-1)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    print("=" * 50)
    print(model.__class__.__name__)
    print("=" * 50)
    print("Accuracy :", accuracy)
    print("Precision:", precision)
    print("Recall   :", recall)
    print("F1 Score :", f1)
    print("\nClassification Report\n")
    print(classification_report(y_test, y_pred))

    return [model.__class__.__name__, accuracy, precision, recall, f1]
```

### Debugging Individual Model Prediction Shapes

Let's verify the prediction shapes of each individual model to identify the culprit, as suggested. We'll specifically check the output shape and type of each model when predicting on the test set.

```python
for name, model in [
    ("RF", rf),
    ("XGB", xgb),
    ("LGBM", lgbm),
    ("CAT", cat)
]:
    # Ensure consistent input type (NumPy array) as used in evaluate_model
    pred = model.predict(X_test.values if isinstance(X_test, pd.DataFrame) else X_test)

    print(f"Estimator: {name}")
    print("Shape:", np.array(pred).shape)
    print("Type:", type(pred))
    print("-"*50)

print("\n--- Detailed XGBoost Prediction Check ---")
pred_xgb = xgb.predict(X_test.values if isinstance(X_test, pd.DataFrame) else X_test)
print("Type:", type(pred_xgb))
print("Shape:", np.array(pred_xgb).shape)
print("First 5 predictions:", pred_xgb[:5])
```

### Fixing XGBoost to Return Class Labels

As anticipated, if `XGBoost` returns a 2D array (probabilities) for multiclass classification, we need to explicitly set its `objective` to `'multi:softmax'` and `num_class` to the number of unique target classes. After making this change, we'll re-fit the `XGBoost` model.

```python
from xgboost import XGBClassifier

# Get the number of unique classes from y_train
num_classes = len(y_train.unique())

xgb = XGBClassifier(
    objective='multi:softmax', # Forces class label output for multiclass
    num_class=num_classes,     # Number of unique classes
    random_state=42,
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6
)

xgb.fit(
    X_train,
    y_train
)

print("Re-trained XGBoost model with 'multi:softmax' objective.")

# Verify XGBoost prediction output shape after correction
print("\n--- Verifying corrected XGBoost predictions ---")
pred_xgb_fixed = xgb.predict(X_test.values if isinstance(X_test, pd.DataFrame) else X_test)
print("Type:", type(pred_xgb_fixed))
print("Shape:", np.array(pred_xgb_fixed).shape)
print("First 5 predictions:", pred_xgb_fixed[:5])
```

### Rebuilding and Re-evaluating the Voting Classifier

Now that the `XGBoost` model is configured to output class labels, we can rebuild the `VotingClassifier` and evaluate its performance. This should resolve the `ValueError`.

```python
from sklearn.ensemble import VotingClassifier

ensemble = VotingClassifier(
    estimators=[
        ('rf', rf),
        ('xgb', xgb),
        ('lgbm', lgbm),
        ('cat', cat)
    ],
    voting='hard'
)

ensemble.fit(
    X_train,
    y_train
)

print("Re-fitted VotingClassifier with corrected XGBoost model.")

results.append(
    evaluate_model(
        ensemble,
        X_test,
        y_test
    )
)

results_df = pd.DataFrame(
    results,
    columns=[
        "Model",
        "Accuracy",
        "Precision",
        "Recall",
        "F1"
    ]
)

print("\n--- Updated Model Comparison ---")
display(results_df.sort_values(
    by="Accuracy",
    ascending=False
))
```
